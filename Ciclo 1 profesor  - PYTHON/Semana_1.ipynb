<center> <h1> Los lenguajes de programación </h1> </center>

</br>

- Un lenguaje de programación es, en su definición más sencilla, un conjunto de palabras reservadas llamadas palabras clave que facilitan esta tarea.

- Mediante los lenguajes de programación podemos elaborar algoritmos (instrucciones) que, una vez *compilados* o *interpretados*, actúan como órdenes para el ordenador.

- Generalmente, escribimos nuestros algoritmos en archivos de texto con una extensión (lo que va después del punto) que depende del lenguaje (.c para C, .py para python, etc.).



<center> <h1> Tipos de lenguajes de programación </h1> </center>

</br>

- Hay centenares de lenguajes de programación, y estos se pueden clasificar (principalmente) de tres formas:
    - De bajo y alto nivel.
    - Compilados o interpretados.
    - Fuerte o débilmente tipados.

Nota: hay muchas más formas de clasificar los lenguajes de programación: campo de aplicación, paradigma funcional vs orientado a objetos, etc. 

<center> <h1> Lenguajes de bajo y alto nivel </h1> </center>

- Los lenguajes de bajo nivel son, por lo general, mucho más complicados y requieren más líneas de código para representar la misma instrucción. Ejemplos: C, Ensamblador.

- Los lenguajes de alto nivel, por contra, tienen una curva de aprendizaje más amigable y requieren menos líneas de código. Ejemplos: Python, Java.

Nota: la interpretación de "bajo" y "alto" nivel es muy subjetiva, para muchas personas solo lenguajes con la complejidad de Ensamblador son de bajo nivel.

<center> <h1> Lenguajes compilados e interpretados </h1> </center>

- Los lenguajes compilados a menudo requieren que un compilador transforme el código (nuestro archivo de texto) en un archivo ejecutable (como los .exe en el caso de Windows). Esta compilación a menudo involucra una optimización del código. Por ej.: C, C++.

- Los lenguajes interpretados se ejecutan bajo un "intérprete" que traduce el código a una versión intermedia en tiempo real y lo ejecuta. A menudo, estos lenguajes de programación son más lentos. Por ej.: Python, Javascript. 

<center> <h1> Lenguajes fuerte y débilmente tipados </h1> </center>

- Los lenguajes fuertemente tipados a menudo exigen que el usuario defina de antemano el tipo de cada variable. Esto resulta en una mayor cantidad de carácteres por línea de código, pero facilita el proceso de compilación y resulta en una ejecución más rápida. Ejemplos: Java, C#.

- Los lenguajes débilmente tipados no requieren que el usuario defina los tipos de las variables. El intérprete se encarga en tiempo real de inferir el tipo de dato de cada variable. Esto tiene un costo computacional que se refleja en la ejecución de los algoritmos. Ejemplos: Python, JavaScript.

<center> <h1>Variables en Python </h1> </center>

- Una variable es el nombre que se le asigna, dentro de un programa, a una posición en memoria. Al tener la dirección de esta posición podemos manipular su contenido.

- Las variables **siempre** deben comenzar por un carácter estándar (a-z, A-Z) y no por un número o un cáracter especial. 

- De preferencia, evitar carácteres como la ñ, pues no son reconocidos en todos los sistemas de codificación de texto.

- Se pueden usar dos convenciones de nombrado de variables: **snake_case** y **camelCase**. Ambas son válidas, pero es preferible usar solo una a la vez.

- Se recomienda utilizar nombres de variable muy simples y no excesivamente largos. Pero estos deben ser dicientes.

- Los nombres de variable NO pueden tener espacios.

- Por lo general, se recomienda tener menos de 80 carácteres por cada línea de código.

- Los comentarios se activan con # o con comillas triples. 

- Solo se usan comentarios si el código no es suficientemente claro de por sí.

<center> <h1> Operaciones de salida de datos </h1> </center>

- La principal operación para salida de datos en Python es **print**.

- Por el momento, la usaremos para mostrar datos por pantalla, pero se puede usar para escribir directamente en archivos también.

- Esta tiene la siguiente forma básica:
    - **print**(lista de variables y mensajes separados por coma)
    
   
Nota: los mensajes deben ir encerrados entre comillas dobles o simples.

```python
print ("Hola")
print ("mundo")
print ("Hola mundo")
print("----------------------")
print ("Hola", "mundo", end = "\n")
print ("Hola", end=" ")
print ("mundo")
print("----------------------")
print ("Hola","jóvenes","cómo","amanecen")
print ("Hola jóvenes cómo amanecen")
print ("Hola","jóvenes","cómo","amanecen", sep=",")
print ("Hola","jóvenes","\ncómo","amanecen", sep=", ")
print("----------------------")

nombre = "Alberto"
print("Mi nombre es ", nombre, "y tengo", 25, "años.")
```

```python
a = 7 + 5
b = a + 2
print(a)
print(7 + 5)
print(f"7 + 5 + 2 = {b}, 7 + 5 = {a}")
print("7 + 5 + 2 = {}, 7 + 5 = {}".format(b, a))
```

```python
cita = ' "Python es un lenguaje de alto nivel" '
print(cita)

cita = " 'C es un lenguaje de bajo nivel' "
print(cita)

cita = " 'Ensamblador es un lenguaje de \"muy\" bajo nivel' "
print(cita)
```
